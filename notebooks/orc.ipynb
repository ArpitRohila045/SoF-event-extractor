```python
img_path = r"C:\Users\hp\Desktop\sof-event-extracter\src\SoF_event_extracter\Dataset\sample2.tif"
img = Image.open(img_path)
draw = ImageDraw.Draw(img)

boxes = pytesseract.image_to_boxes(img)
w, h = img.size

for b in boxes.splitlines():
    b = b.split(' ')
    # b[1:5] are x1, y1, x2, y2
    x1, y1, x2, y2 = int(b[1]), int(b[2]), int(b[3]), int(b[4])
    # Tesseract's y-coordinates are from bottom, PIL's from top
    draw.rectangle([x1, h - y2, x2, h - y1], outline='red', width=2)

img.show()
```
